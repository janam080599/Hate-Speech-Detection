```
train_df=pd.read_csv('/content/drive/My Drive/Colab_Notebooks/Datasets/hatespeech_kaggle/train.csv')
train_df.head(100)
```

```
test_df=pd.read_csv('/content/drive/My Drive/Colab_Notebooks/Datasets/hatespeech_kaggle/test.csv')
test_df.head(100)
```

```
nltk.download('stopwords')
```

```
from nltk.stem import WordNetLemmatizer
from nltk import tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import re
```

```
train_df['text_lem'] = [''.join([WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]',' ',text)) for text in lis]) for lis in train_df['tweet']]
test_df['text_lem'] = [''.join([WordNetLemmatizer().lemmatize(re.sub('[^A-Za-z]',' ',text)) for text in lis]) for lis in test_df['tweet']]
```

```
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(train_df['text_lem'],train_df['label'])
```

```
vect = TfidfVectorizer(ngram_range = (1,4)).fit(X_train)
```

```
vect_transformed_X_train = vect.transform(X_train)
vect_transformed_X_test = vect.transform(X_test)
```

```
modelNB = MultinomialNB(alpha=1.7).fit(vect_transformed_X_train,y_train)
```

```
predictionsNB = modelNB.predict(vect_transformed_X_test)
sum(predictionsNB==1),len(y_test),f1_score(y_test,predictionsNB)
```

```
accuracy_score(predictionsNB,y_test)
```

```
print('The confusion matrix for the model is')
print(confusion_matrix(y_test,predictionsNB.round()))
```
